```python
import pandas as pd
import numpy as np
```

Import our white and red wine data from the csvs

# Read the CSV and Perform Basic Data Cleaning

```python
red_wine_df = pd.read_csv("Resources/winequality-red.csv", sep=';')
white_wine_df = pd.read_csv("Resources/winequality-white.csv", sep=';')
white_wine_df.info()
red_wine_df.info()
```

## Unbalanced (Denormalized) Data for White Wine Data (4898 data points) and Red Wine Data (1599 data points) 

We want to use all the white and red wine data baseline condition we can use:

```python
#if we want our only features of the model to consist of all white and red wine data, we use this data frame:
baseline_red_df = red_wine_df.copy()
baseline_white_df = white_wine_df.copy()
baseline_wine_df = pd.concat([baseline_red_df, baseline_white_df])
```

```python
baseline_wine_df.groupby('quality').mean()
```

```python
#Suppose we want to take specific ranges in only_balanced_df, which does not include colors
# range_colors_df = dummy_colors_df.loc[\
# (dummy_colors_df["quality"] <= 8 ) & \
# (dummy_colors_df["quality"] >= 3 )  \
# ] 
```

```python
bin_balanced_df = baseline_wine_df.copy()
bins = [0, 4, 5, 6, 10]
group_names = [0, 1, 2, 3]
bin_balanced_df["bin_quality"] = pd.cut(bin_balanced_df["quality"], bins, labels=group_names)
bin_balanced_df.head(30)
```

```python
bin_balanced_df.groupby("bin_quality").count()
```

```python
bad_df = bin_balanced_df.loc[bin_balanced_df['bin_quality']==0,:]
good_df = bin_balanced_df.loc[bin_balanced_df['bin_quality']==1,:]
great_df = bin_balanced_df.loc[bin_balanced_df['bin_quality']==2,:]
terrific_df = bin_balanced_df.loc[bin_balanced_df['bin_quality']==3,:]
```

```python
normalized_good_df = good_df.sample(n=246,random_state=42)
```

```python
normalized_great_df = great_df.sample(n=246,random_state=42)
```

```python
normalized_terrific_df = terrific_df.sample(n=246,random_state=42)
```

```python
normal_df = pd.concat([bad_df,normalized_good_df,normalized_great_df,normalized_terrific_df])
normdf = normal_df.drop(columns = 'quality')
normdf
```

# Select your features (columns)

```python
# Set features. Drop quality, y value. 
X = normdf.drop(columns=['bin_quality'])
X.head()
```

# Create a Train Test Split

Use `quality` for the y values

```python
from sklearn.model_selection import train_test_split
#80% train, 20% test. y is this one column
y = normdf[['bin_quality']]
#random state 42 will have same picks for x test and y test
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

# Train the Model using Random Forest 



```python
#Train the model
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
#Test the model
from sklearn.metrics import confusion_matrix, classification_report
predictions = rfc.predict(X_test)
print(classification_report(y_test, predictions))
print(predictions[9:20])
print(y_test[9:20])
```
